```python
import numpy as np
import HMM
import importlib
```

```python
word_list = []
dictionary = set()
with open('data/Syllable_dictionary.txt') as f:
    for line in f:
        word_list.append(line.split()[0])
        dictionary.add(line.split()[0])
```

```python
def remove_punctuation(words):
    punctuation = [',','.',':','?',';','!',"'",'"', '(', ')']
    for i, word in enumerate(words):
        word = word.lower()

        while word not in dictionary:
            
            if word[-1] in punctuation:
                word = word[:-1]

            if word[0] in punctuation and word not in dictionary:
                word = word[1:]

        words[i] = word
            
    return words
```

```python
def load_data(filename):
    data = []
    with open(filename) as f:
        start = 0
        current = []
        for i, line in enumerate(f):
            words = line.split()
            if len(words) < 1:
                continue

            if len(words) == 1:
                start = i
                continue

            words = remove_punctuation(words)
            current.extend(words) 

            if i == start + 4 or i == start + 8 or i == start + 12 or \
            i == start + 14:
                data.append(current)
                current = []

    return data
```

```python
def encode_data_HMM(data):
    encoding = {}
    for i, word in enumerate(word_list):
        encoding[word] = i
    
    encoded_data = []
    for x in data:
        encoded_x = []
        for word in x:
            encoded_x.append(encoding[word])
        encoded_data.append(encoded_x)
        
    return encoded_data
```

```python
data = load_data('data/shakespeare.txt')
data_HMM = encode_data_HMM(data)
print(len(data_HMM))
```
